```python
from diffusers import StableDiffusionPipeline
import torch

def load_model():
    model_id = "nota-ai/bk-sdm-tiny"
    pipe = StableDiffusionPipeline.from_pretrained(
        model_id,
        torch_dtype=torch.float16 if torch.cuda.is_available() else torch.float32
    )
    device = "cuda" if torch.cuda.is_available() else "cpu"
    pipe = pipe.to(device)
    return pipe

pipe = load_model()
print("Model loaded!")
```

```python
def enhance_fashion_prompt(user_prompt):
    fashion_prefix = (
        "fashion design, clothing sketch, high quality, detailed, "
        "professional fashion illustration, white background, "
    )
    fashion_suffix = (
        ", studio lighting, fashion week style, clean lines, "
        "fabric texture visible, full outfit view"
    )
    return fashion_prefix + user_prompt + fashion_suffix

# Test it
raw = "a flowy summer dress with floral print"
enhanced = enhance_fashion_prompt(raw)
print(enhanced)
```

```python
from PIL import Image
import os

def generate_design(pipe, prompt, num_variants=3, output_dir="outputs"):
    os.makedirs(output_dir, exist_ok=True)
    enhanced = enhance_fashion_prompt(prompt)
    images = []
    
    for i in range(num_variants):
        image = pipe(
                    prompt,
                    num_inference_steps=15,   # reduce from 30 → still decent quality
                    guidance_scale=7.5
                ).images[0]
        
        path = f"{output_dir}/design_{i+1}.png"
        image.save(path)
        images.append(path)
        print(f"Generated variant {i+1}")
    
    return images

# Test generation
paths = generate_design(pipe, "a flowy summer dress with floral print")
```

```python
from sentence_transformers import SentenceTransformer
import numpy as np

# Sample product catalogue (replace with real API later)
PRODUCT_CATALOGUE = [
    {"name": "Floral Midi Dress", "price": "₹799", "brand": "FabIndia", "url": "#", "tags": "floral summer dress flowy"},
    {"name": "Printed Wrap Dress", "price": "₹1,199", "brand": "W for Woman", "url": "#", "tags": "printed wrap casual"},
    {"name": "Boho Sundress", "price": "₹649", "brand": "Aurelia", "url": "#", "tags": "boho floral summer"},
    {"name": "Denim Jacket", "price": "₹999", "brand": "Roadster", "url": "#", "tags": "denim jacket casual"},
    {"name": "Embroidered Kurti", "price": "₹549", "brand": "Biba", "url": "#", "tags": "embroidered ethnic cotton"},
]

embed_model = SentenceTransformer("all-MiniLM-L6-v2")

def find_similar_products(user_prompt, top_k=3):
    query_emb = embed_model.encode(user_prompt)
    scores = []
    for product in PRODUCT_CATALOGUE:
        prod_emb = embed_model.encode(product["tags"])
        score = np.dot(query_emb, prod_emb) / (
            np.linalg.norm(query_emb) * np.linalg.norm(prod_emb)
        )
        scores.append((score, product))
    scores.sort(reverse=True)
    return [p for _, p in scores[:top_k]]

# Test
results = find_similar_products("floral summer dress")
for r in results:
    print(r["name"], r["price"])
```
